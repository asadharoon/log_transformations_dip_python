# Import Libraries

```python
from skimage import data
from scipy.misc import imread, imresize
import numpy as np
from scipy import ndimage
import matplotlib.pyplot as plt
```

# Log Transformation

Formula of Log Transformation is:
s =c*Log(1+r)

# Original Image

```python
original_image =imread('test2.png',True, 'L')    #read image as grey scale image
```

# Processed Image

```python
processed_img =original_image.copy()
```

```python
c=1
processed_img=c*np.log(1+original_image)
#print(np.amin(processed_img),np.amax(processed_img))
```

```python
processed_img[processed_img<0]=0
processed_img[processed_img>=255]=255
```

```python
fig, axes = plt.subplots(1, 2)
ax = axes.ravel()

ax[0].imshow(original_image,cmap=plt.cm.gray,interpolation='bilinear')
ax[0].set_title("Original")
ax[1].imshow(processed_img, cmap=plt.cm.gray,interpolation='bilinear')
ax[1].set_title("Log Transformations")
plt.imshow(processed_img,cmap=plt.cm.gray) 
```
